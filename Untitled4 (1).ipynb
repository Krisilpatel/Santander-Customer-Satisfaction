```python
## The target variable is rare event where 96% values are 0 and 4% are 1 
## Also to improve performance, redunduncy from data has to be removed. For that constant and equal attributes is dropped.
```

```python
## Removing Constant Attributes:

no_unique=train.apply(lambda x: len(x.unique()))
train.drop(no_unique[no_unique==1].index,inplace=True, axis=1)
test.drop(no_unique[no_unique==1].index,inplace=True, axis=1)
```

```python
## Removing Same Attributes:
from itertools import combinations
combi = list(combinations(train.columns.tolist(),2))
equal_comb=[]
for i in combi:
    same=np.array_equal(train[i[0]],train[i[1]])
    if same:
        equal_comb.append(list(i))
```

```python
train.drop(np.array(equal_comb)[:,1],axis=1, inplace=True)


```

```python
cols=train.columns.tolist()
cols.remove('TARGET')
X=train[cols]
Y=train['TARGET']
```

```python
from sklearn.model_selection import cross_validate
from sklearn import tree
from sklearn import metrics
from sklearn import ensemble
from sklearn import linear_model 
from sklearn import naive_bayes 
import xgboost as xgb

skf = cross_validate.StratifiedKFold(Y, n_folds=5, shuffle=True)
score_metric = 'roc_auc'
scores = {}
```

```python
scores['DicisionTree']=cross_validation.cross_val_score(tree.DecisionTreeClassifier(), X, Y, cv=skf, scoring=score_metric)
scores['RandomForest']=cross_validation.cross_val_score(ensemble.RandomForestClassifier(), X, Y, cv=skf, scoring=score_metric)
scores['Ada_Boost']=cross_validation.cross_val_score(ensemble.AdaBoostClassifier(), X, Y, cv=skf, scoring=score_metric)
scores['xgboost']=cross_validation.cross_val_score(xgb.XGBClassifier(), X, Y, cv=skf, scoring=score_metric)
scores['ridge']=cross_validation.cross_val_score(linear_model.RidgeClassifier(), X, Y, cv=skf, scoring=score_metric)
scores['gradient_boost']=cross_validation.cross_val_score(ensemble.GradientBoostingClassifier(), X, Y, cv=skf, scoring=score_metric)
scores['bagging'] = cross_validation.cross_val_score(ensemble.BaggingClassifier(), X, Y, cv=skf, scoring=score_metric)
```

```python
model_scores = pd.DataFrame(scores)
```
